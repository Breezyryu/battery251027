```python
# í™˜ê²½ ì„¤ì •
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# í•œê¸€ í°íŠ¸
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# BatteryDataTool ëª¨ë“ˆ Import
sys.path.append(r'c:\Users\Ryu\Python_project\data\battery251027')
from BatteryDataTool import (
    check_cycler, name_capacity,
    pne_min_cap, pne_cycle_data
)

print("âœ… í™˜ê²½ ì„¤ì • ì™„ë£Œ")
```

```python
# ë°ì´í„° ê²½ë¡œ ì„¤ì •
BASE_PATH = r"C:\Users\Ryu\Python_project\data\BatteryData_251010\Rawdata"

# ì—°ì† ê²½ë¡œ ë¦¬ìŠ¤íŠ¸ (ìˆœì„œëŒ€ë¡œ)
folder_paths = [
    os.path.join(BASE_PATH, "A1_MP1_4500mAh_T23_1"),
    os.path.join(BASE_PATH, "A1_MP1_4500mAh_T23_2"),
    os.path.join(BASE_PATH, "A1_MP1_4500mAh_T23_3")
]

# ê²½ë¡œ í™•ì¸
for i, path in enumerate(folder_paths, 1):
    exists = "âœ…" if os.path.exists(path) else "âŒ"
    print(f"{exists} í´ë” {i}: {os.path.basename(path)}")
    if os.path.exists(path):
        cycler_type = "PNE" if check_cycler(path) == True else "TOYO" if check_cycler(path) == False else "UNKNOWN"
        capacity = name_capacity(os.path.basename(path))
        print(f"      ì¶©ë°©ì „ê¸°: {cycler_type}, ìš©ëŸ‰: {capacity} mAh")
```

```python
# ê° í´ë”ì—ì„œ ì‚¬ì´í´ ë°ì´í„° ë¡œë”©
print("\n=== ì‚¬ì´í´ ë°ì´í„° ë¡œë”© ì¤‘ ===\n")

ini_crate = 0.2  # 0.2C rate
chkir = True
chkir2 = False
mkdcir = True

cycle_data_list = []

for i, folder_path in enumerate(folder_paths, 1):
    if not os.path.exists(folder_path):
        print(f"âš ï¸ í´ë” {i}: ê²½ë¡œ ì—†ìŒ, ê±´ë„ˆëœ€")
        continue
    
    try:
        print(f"ğŸ“‚ í´ë” {i}: {os.path.basename(folder_path)}")
        
        # ìš©ëŸ‰ ìë™ ì‚°ì •
        capacity = pne_min_cap(folder_path, ini_crate)
        print(f"   ìš©ëŸ‰: {capacity} mAh (ìë™ ì‚°ì •)")
        
        # PNE ë°ì´í„° ë¡œë”©
        cycle_df = pne_cycle_data(folder_path, capacity, ini_crate, chkir, chkir2, mkdcir)
        
        if cycle_df is not None and len(cycle_df) > 0:
            cycle_data_list.append(cycle_df)
            print(f"   âœ… ë¡œë”© ì™„ë£Œ: {len(cycle_df)} ì‚¬ì´í´")
            print(f"   ì‚¬ì´í´ ë²”ìœ„: {cycle_df.index.min()} ~ {cycle_df.index.max()}")
        else:
            print(f"   âš ï¸ ë°ì´í„° ì—†ìŒ")
    
    except Exception as e:
        print(f"   âŒ ì˜¤ë¥˜: {e}")

print(f"\nâœ… ì´ {len(cycle_data_list)}ê°œ í´ë” ë¡œë”© ì™„ë£Œ")
```

```python
# ë°ì´í„° í†µí•©
if len(cycle_data_list) > 0:
    # ëª¨ë“  DataFrame ì—°ê²°
    combined_df = pd.concat(cycle_data_list, axis=0)
    
    # ì¤‘ë³µ ì¸ë±ìŠ¤ ì œê±°
    combined_df = combined_df[~combined_df.index.duplicated(keep='first')]
    
    # ì¸ë±ìŠ¤ ì •ë ¬
    combined_df = combined_df.sort_index()
    
    print(f"\n=== í†µí•© ë°ì´í„° ì •ë³´ ===\n")
    print(f"ì´ ì‚¬ì´í´ ìˆ˜: {len(combined_df)}")
    print(f"ì‚¬ì´í´ ë²”ìœ„: {combined_df.index.min()} ~ {combined_df.index.max()}")
    print(f"\në°ì´í„° ì»¬ëŸ¼: {list(combined_df.columns)}")
    print(f"\në°ì´í„° ë¯¸ë¦¬ë³´ê¸°:")
    print(combined_df.head(10))
else:
    print("âš ï¸ ë¡œë”©ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
```

```python
# ì‚¬ì´í´ ë¶„ì„ ê·¸ë˜í”„
if len(cycle_data_list) > 0:
    # í†µí•© ë°ì´í„° ìš©ëŸ‰ ê³„ì‚° (ì²« ë²ˆì§¸ í´ë” ê¸°ì¤€)
    if len(folder_paths) > 0 and os.path.exists(folder_paths[0]):
        display_capacity = pne_min_cap(folder_paths[0], ini_crate)
    else:
        display_capacity = 4500
    
    fig, axes = plt.subplots(3, 2, figsize=(15, 12))
    fig.suptitle(f'A1 MP1 {display_capacity}mAh T23 ì—°ì† ë¶„ì„ (PNE)', 
                 fontsize=16, fontweight='bold')
    
    cycles = combined_df.index.values
    
    # 1. ë°©ì „ ìš©ëŸ‰
    axes[0, 0].plot(cycles, combined_df['Dchg'], 'o-', color='#2E86AB', linewidth=2, markersize=3)
    axes[0, 0].set_xlabel('Cycle', fontsize=12)
    axes[0, 0].set_ylabel('ë°©ì „ ìš©ëŸ‰ ë¹„ìœ¨', fontsize=12)
    axes[0, 0].set_title('1. ë°©ì „ ìš©ëŸ‰ ë³€í™”', fontsize=12, fontweight='bold')
    axes[0, 0].set_ylim(0.70, 1.05)
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].axhline(y=0.8, color='r', linestyle='--', linewidth=1, alpha=0.5, label='80% ê¸°ì¤€')
    axes[0, 0].legend()
    
    # 2. ì¶©ì „ ìš©ëŸ‰
    axes[0, 1].plot(cycles, combined_df['Chg'], 'o-', color='#A23B72', linewidth=2, markersize=3)
    axes[0, 1].set_xlabel('Cycle', fontsize=12)
    axes[0, 1].set_ylabel('ì¶©ì „ ìš©ëŸ‰ ë¹„ìœ¨', fontsize=12)
    axes[0, 1].set_title('2. ì¶©ì „ ìš©ëŸ‰ ë³€í™”', fontsize=12, fontweight='bold')
    axes[0, 1].set_ylim(0.70, 1.05)
    axes[0, 1].grid(True, alpha=0.3)
    
    # 3. ì¶©ë°© íš¨ìœ¨
    axes[1, 0].plot(cycles, combined_df['Eff'], 'o-', color='#F18F01', linewidth=2, markersize=3)
    axes[1, 0].set_xlabel('Cycle', fontsize=12)
    axes[1, 0].set_ylabel('íš¨ìœ¨', fontsize=12)
    axes[1, 0].set_title('3. ì¶©ë°© íš¨ìœ¨ (Dchg/Chg)', fontsize=12, fontweight='bold')
    axes[1, 0].set_ylim(0.95, 1.02)
    axes[1, 0].grid(True, alpha=0.3)
    
    # 4. ì˜¨ë„
    axes[1, 1].plot(cycles, combined_df['Temp'], 'o-', color='#BC4B51', linewidth=2, markersize=3)
    axes[1, 1].set_xlabel('Cycle', fontsize=12)
    axes[1, 1].set_ylabel('ì˜¨ë„ (Â°C)', fontsize=12)
    axes[1, 1].set_title('4. ì˜¨ë„ ë³€í™”', fontsize=12, fontweight='bold')
    axes[1, 1].grid(True, alpha=0.3)
    
    # 5. DCIR
    if 'dcir' in combined_df.columns:
        axes[2, 0].plot(cycles, combined_df['dcir'], 'o-', color='#6A994E', linewidth=2, markersize=3)
        axes[2, 0].set_xlabel('Cycle', fontsize=12)
        axes[2, 0].set_ylabel('DCIR (mÎ©)', fontsize=12)
        axes[2, 0].set_title('5. DCIR ë³€í™”', fontsize=12, fontweight='bold')
        axes[2, 0].grid(True, alpha=0.3)
    
    # 6. í´ë”ë³„ í‰ê·  ë¹„êµ
    folder_labels = ['Folder 1', 'Folder 2', 'Folder 3']
    folder_avg_cap = []
    
    for df in cycle_data_list:
        folder_avg_cap.append(df['Dchg'].mean())
    
    axes[2, 1].bar(folder_labels[:len(folder_avg_cap)], folder_avg_cap, 
                   color=['#2E86AB', '#A23B72', '#F18F01'])
    axes[2, 1].set_xlabel('í´ë”', fontsize=12)
    axes[2, 1].set_ylabel('í‰ê·  ë°©ì „ ìš©ëŸ‰ ë¹„ìœ¨', fontsize=12)
    axes[2, 1].set_title('6. í´ë”ë³„ í‰ê·  ìš©ëŸ‰', fontsize=12, fontweight='bold')
    axes[2, 1].set_ylim(0.70, 1.05)
    axes[2, 1].grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.show()
```

```python
# í†µê³„ ë¶„ì„
if len(cycle_data_list) > 0:
    print("\n" + "="*80)
    print("ìƒì„¸ í†µê³„ ë¶„ì„")
    print("="*80 + "\n")
    
    # ìš©ëŸ‰ ê³„ì‚°
    if len(folder_paths) > 0 and os.path.exists(folder_paths[0]):
        capacity = pne_min_cap(folder_paths[0], ini_crate)
    else:
        capacity = 4500
    
    print(f"ğŸ“Š ì „ì²´ ë°ì´í„°")
    print(f"   ì´ ì‚¬ì´í´: {len(combined_df)}")
    print(f"   ì‚¬ì´í´ ë²”ìœ„: {combined_df.index.min()} ~ {combined_df.index.max()}")
    print(f"   ìš©ëŸ‰: {capacity} mAh")
    
    print(f"\nğŸ’¾ ë°©ì „ ìš©ëŸ‰")
    initial_cap = combined_df['Dchg'].iloc[0]
    final_cap = combined_df['Dchg'].iloc[-1]
    print(f"   ì´ˆê¸°: {initial_cap:.4f} ({initial_cap * capacity:.1f} mAh)")
    print(f"   ìµœì¢…: {final_cap:.4f} ({final_cap * capacity:.1f} mAh)")
    print(f"   ìš©ëŸ‰ ìœ ì§€ìœ¨: {(final_cap / initial_cap) * 100:.2f}%")
    print(f"   ìš©ëŸ‰ ì†ì‹¤: {((initial_cap - final_cap) / initial_cap) * 100:.2f}%")
    
    # 80% ë„ë‹¬ ì‚¬ì´í´ ì°¾ê¸°
    below_80 = combined_df[combined_df['Dchg'] < 0.8]
    if len(below_80) > 0:
        eol_cycle = below_80.index[0]
        print(f"   80% EOL ë„ë‹¬: Cycle {eol_cycle}")
    else:
        print(f"   80% EOL: ë¯¸ë„ë‹¬")
    
    print(f"\nâš¡ íš¨ìœ¨")
    print(f"   í‰ê·  ì¶©ë°© íš¨ìœ¨: {combined_df['Eff'].mean():.4f} ({combined_df['Eff'].mean() * 100:.2f}%)")
    print(f"   ìµœì†Œ íš¨ìœ¨: {combined_df['Eff'].min():.4f}")
    print(f"   ìµœëŒ€ íš¨ìœ¨: {combined_df['Eff'].max():.4f}")
    
    print(f"\nğŸŒ¡ï¸ ì˜¨ë„")
    print(f"   í‰ê·  ì˜¨ë„: {combined_df['Temp'].mean():.2f} Â°C")
    print(f"   ìµœì € ì˜¨ë„: {combined_df['Temp'].min():.2f} Â°C")
    print(f"   ìµœê³  ì˜¨ë„: {combined_df['Temp'].max():.2f} Â°C")
    print(f"   ì˜¨ë„ í¸ì°¨: {combined_df['Temp'].std():.2f} Â°C")
    
    if 'dcir' in combined_df.columns:
        print(f"\nğŸ”‹ DCIR")
        print(f"   í‰ê·  DCIR: {combined_df['dcir'].mean():.2f} mÎ©")
        print(f"   ì´ˆê¸° DCIR: {combined_df['dcir'].iloc[0]:.2f} mÎ©")
        print(f"   ìµœì¢… DCIR: {combined_df['dcir'].iloc[-1]:.2f} mÎ©")
        print(f"   DCIR ì¦ê°€ìœ¨: {((combined_df['dcir'].iloc[-1] / combined_df['dcir'].iloc[0]) - 1) * 100:.2f}%")
    
    print(f"\nğŸ“ˆ í´ë”ë³„ ë¶„ì„")
    for i, df in enumerate(cycle_data_list, 1):
        print(f"\n   í´ë” {i}:")
        print(f"      ì‚¬ì´í´ ìˆ˜: {len(df)}")
        print(f"      ì‚¬ì´í´ ë²”ìœ„: {df.index.min()} ~ {df.index.max()}")
        print(f"      í‰ê·  ë°©ì „ ìš©ëŸ‰: {df['Dchg'].mean():.4f}")
        print(f"      í‰ê·  íš¨ìœ¨: {df['Eff'].mean():.4f}")
        print(f"      í‰ê·  ì˜¨ë„: {df['Temp'].mean():.2f} Â°C")
        if 'dcir' in df.columns:
            print(f"      í‰ê·  DCIR: {df['dcir'].mean():.2f} mÎ©")
```

```python
# Excel ì €ì¥ (ì˜µì…˜)
save_excel = True

if save_excel and len(cycle_data_list) > 0:
    output_path = os.path.join(BASE_PATH, "A1_MP1_4500mAh_T23_ì—°ì†_í†µí•©ë¶„ì„.xlsx")
    
    with pd.ExcelWriter(output_path, engine='xlsxwriter') as writer:
        # í†µí•© ë°ì´í„°
        combined_df.to_excel(writer, sheet_name='í†µí•©_ì‚¬ì´í´_ë°ì´í„°')
        
        # ê°œë³„ í´ë” ë°ì´í„°
        for i, df in enumerate(cycle_data_list, 1):
            df.to_excel(writer, sheet_name=f'í´ë”_{i}_ë°ì´í„°')
        
        # í†µê³„
        if len(folder_paths) > 0 and os.path.exists(folder_paths[0]):
            capacity = pne_min_cap(folder_paths[0], ini_crate)
        else:
            capacity = 4500
        
        stats_data = {
            'í•­ëª©': ['ì´ ì‚¬ì´í´ ìˆ˜', 'ì´ˆê¸° ìš©ëŸ‰ ë¹„ìœ¨', 'ìµœì¢… ìš©ëŸ‰ ë¹„ìœ¨', 'ìš©ëŸ‰ ìœ ì§€ìœ¨ (%)', 
                   'í‰ê·  íš¨ìœ¨', 'í‰ê·  ì˜¨ë„ (Â°C)', 'ìš©ëŸ‰ (mAh)'],
            'ê°’': [
                len(combined_df),
                f"{combined_df['Dchg'].iloc[0]:.4f}",
                f"{combined_df['Dchg'].iloc[-1]:.4f}",
                f"{(combined_df['Dchg'].iloc[-1] / combined_df['Dchg'].iloc[0]) * 100:.2f}",
                f"{combined_df['Eff'].mean():.4f}",
                f"{combined_df['Temp'].mean():.2f}",
                capacity
            ]
        }
        stats_df = pd.DataFrame(stats_data)
        stats_df.to_excel(writer, sheet_name='í†µê³„', index=False)
    
    print(f"\nâœ… Excel ì €ì¥ ì™„ë£Œ: {output_path}")
```
